```python
import re

# Define the regex pattern for the incorrect characters
encoding_issues_pattern = r'[ÃÂâ€™â€œâ€¢Ã±Ã©Â®Ã¼Ã¶â€•Ã¤â€“Ã³ÃŸÃºÃ¡]'

# Function to detect rows with wrong encoding in a string
def detect_wrong_encoding(text):
    if isinstance(text, str):
        return bool(re.search(encoding_issues_pattern, text))  # Returns True if wrong encoding is found
    return False

# Apply the detection function to each column and create a mask for rows with encoding issues
wrong_encoded_rows = raw_df.applymap(detect_wrong_encoding).any(axis=1)

# Extract rows with wrongly encoded data
incorrectly_encoded_df = raw_df[wrong_encoded_rows]

# Display the rows with wrong encoding
print(incorrectly_encoded_df)
```

```python
import ftfy

# Function to fix text encoding issues
def fix_text_encoding(text):
    return ftfy.fix_text(text)

# Apply it to each column in the DataFrame
for col in raw_df.columns:
    raw_df[col] = raw_df[col].apply(lambda x: fix_text_encoding(x) if isinstance(x, str) else x)

# Display the updated DataFrame
print(raw_df.head(20))
```

```python
encoding_issues_pattern = r'[ÃÂâ€™â€œâ€¢Ã±Ã©Â®Ã¼Ã¶â€•Ã¤â€“Ã¶ÃŸÃ³Ã¤Â¿ÃºÃ¡]'

# Detect rows with encoding issues, ignoring NaN values
incorrectly_encoded_titles = raw_df[raw_df['TITLE'].notna() & raw_df['TITLE'].str.contains(encoding_issues_pattern, regex=True, na=False)]
incorrectly_encoded_author = raw_df[raw_df['AUTHOR'].notna() & raw_df['AUTHOR'].str.contains(encoding_issues_pattern, regex=True, na=False)]
incorrectly_encoded_publisher = raw_df[raw_df['PUBLISHER'].notna() & raw_df['PUBLISHER'].str.contains(encoding_issues_pattern, regex=True, na=False)]

# Combine all the problematic rows (only those with encoding issues, not NaN)
rows_to_drop = pd.concat([incorrectly_encoded_titles, incorrectly_encoded_author, incorrectly_encoded_publisher])

# Drop the rows with encoding issues from the original DataFrame
raw_df_cleaned2 = raw_df.drop(rows_to_drop.index)

# Display info about the cleaned DataFrame
print(raw_df_cleaned2.info())

# Display the first 20 rows of the cleaned DataFrame for verification
print(raw_df_cleaned2.head(20))
```
